```python
# Load dataset
from sklearn.datasets import load_digits
dt = load_digits(n_class=3)
```

```python
X = dt.images
Y = dt.target
```

```python
X = X.reshape(537,1,8,8)
```

```python
import torch
import torch.nn as nn
from torch.optim import Adam
from torch.utils.data import TensorDataset, DataLoader
```

```python
# Convert to tensors
X = torch.FloatTensor(X)
Y = torch.LongTensor(Y)
```

```python
# ✅ CREATE DATASET
dataset = TensorDataset(X, Y)
```

```python
loader = DataLoader(dataset, batch_size=32, shuffle=True)
```

```python
model = nn.Sequential(
    nn.Conv2d(1, 32, kernel_size=3, stride=1, padding=1),
    nn.ReLU(),
    nn.MaxPool2d(kernel_size=2, stride=2),

    nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1),
    nn.ReLU(),
    nn.MaxPool2d(kernel_size=2, stride=2),

    nn.Flatten(),
    nn.Linear(64*2*2, 128),
    nn.ReLU(),
    nn.Linear(128, 3)
)
```

```python
# Loss & Optimizer
loss = nn.CrossEntropyLoss()
opt = Adam(model.parameters(), lr=0.001)
```

```python
# ✅ TRAINING WITH BATCHES
for epoch in range(100):
    total_loss = 0

    for xb, yb in loader:   # ✅ BATCH LOOP

        opt.zero_grad()

        yp = model(xb)      # forward pass (batch)
        ls = loss(yp, yb)   # batch loss

        ls.backward()
        opt.step()

        total_loss += ls.item()

    if (epoch+1) % 10 == 0:
        print("Epoch:", epoch+1, "Loss:", total_loss)
```

```python
# Prediction
newx = X[112].reshape(1,1,8,8)

Yp = model(newx)
sx = torch.softmax(Yp, dim=1)

print("Prediction:", torch.argmax(sx, dim=1))
```
